```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('D:/train.csv')

# Display the first few rows of the dataframe
print(df.head())

# Check for missing values
print(df.isnull().sum())

# Fill missing values in Age with the median age
df['Age'].fillna(df['Age'].median(), inplace=True)

# Drop the Cabin column as it has too many missing values
df = df.drop('Cabin', axis=1)

# Fill missing values in Embarked with the mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Check the cleaned data
print(df.isnull().sum())

# Perform EDA

# Plot a histogram of Age
plt.figure(figsize=(10,6))
sns.histplot(df['Age'], kde=False, bins=30)
plt.title('Distribution of Age')
plt.show()

# Plot a bar chart of survival status count
sns.countplot(x='Survived', data=df)
plt.title('Survival Status Count')
plt.show()

# Plot a bar chart of Class count
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Count')
plt.show()

# Explore relationship between variables

# Plot a bar chart of survival count by passenger class
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title('Survival Count by Passenger Class')
plt.show()

# Plot a boxplot of age by passenger class
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age by Passenger Class')
plt.show()
```
